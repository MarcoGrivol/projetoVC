```python
import cv2
import numpy as np
from os import listdir
from os.path import join, isfile
from collections import OrderedDict
import matplotlib.pyplot as plt

PATH = 'C:\\Users\\marco\\My Drive\\ComputerScience\\Vis√£o Computacional\\images\\Skol'

class VisaoComputacional:
    def __init__(self, folder_path, ids):
        self.imgs = []
        self.labels = []
        for i in ids:
            img_with_names = self._getImagesFromFolder(join(folder_path, i))
            self.imgs.extend(img_with_names[0])
            self.labels.extend(img_with_names[1])
        
    def _getImagesFromFolder(self, folder_path):
        imgs = []
        labels = []
        for file in listdir(folder_path):
            img_path = join(folder_path, file)
            if isfile(img_path) and file.endswith('.jpg'):
                img = plt.imread(img_path)
                imgs.append(img)
                labels.append(file)
        return (imgs, labels)
```

```python
vc = VisaoComputacional(PATH, ['0', '1'])
```

```python
query_img = plt.imread(join(PATH, '0\\0_v0_0.jpg'))
train_img = plt.imread(join(PATH, '1\\1_v0_0.jpg'))

fig, axs = plt.subplots(1, 2, figsize=(12, 12))
axs[0].imshow(query_img)
axs[0].set_title('Query img')
axs[1].imshow(train_img)
axs[1].set_title('Train img')
plt.show()
```
